# **‚≠ê Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# view data
disease.head()
```

```python
# info of data
disease.info(verbose=True, null_counts=True)
```

```python
# summary statistics
disease.describe()
```

```python
# check for missing value
disease.isna().sum()
```

```python
# check for categories
disease.nunique()
```

```python
# correlation
disease.corr()
```

```python
# visualize pairplot 
sns.pairplot(disease, vars=['itching', 'skin_rash', 'nodal_skin_eruptions', 'continuous_sneezing',
       'shivering', 'chills', 'joint_pain', 'stomach_pain', 'acidity',
       'ulcers_on_tongue'])
```

```python
# column names
disease.columns
```

```python
# define y
y = disease['prognosis']
```

```python
# define X
X = disease.drop(['prognosis'],axis=1)
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2529)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
# verify shape
disease.shape
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# train model
model.fit(X_train, y_train)
```

```python
# predict with model
y_pred = model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
# model accuracy
accuracy_score(y_test, y_pred)
```

```python
# model confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# model classification report
print(classification_report(y_test,y_pred))
```
